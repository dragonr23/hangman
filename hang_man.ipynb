```python
import random
from IPython.display import clear_output


#Lists of Potential Words
animals = ['lizard', 'zebra', 'giraffe','elephant','lion','wolf','rhinoceros','monkey','lemar','geiko']
lakes = ['winnipesaukee', 'opechee', 'winnisquam', 'sunapee','waukewan','newfound','merrymeeting','pawtuckaway','kanasatka','pemigewasset']

#Hangman Functions
def hangMan():
    lives = [5]
    letters = []
    def lifeLost(lives):
        lives[0] = lives[0] - 1
        return lives

    def guessedLetter(no_guesses, random_word):
        while True:
            #if random_word = no_guesses = CONGRATS break
            if random_word == no_guesses.replace(" ",''):
                print('Congrats! You won hangman!')
                break
            elif lives[0] == 0:
                print("YOU'RE A DEAD MAN")
                play_again = input("Would you like to play again?")
                if play_again.lower() == 'yes':
                    clear_output()
                    lives[0] = 5
                    hangMan()
                else:
                    break
            #else:
            else:
                guess = input("What letter would you like to guess?   ")
                letters.append(guess)
                no_guesses_list = list(no_guesses)
                random_word_list = list(random_word)
                guessed_string = no_guesses
                clear_output()

                if guess in random_word:

                    for i, x in enumerate(random_word_list):
                        if x == guess:
                            no_guesses_list[i + i] = random_word_list[i]
                            guessed_string = ''.join(no_guesses_list)
                            no_guesses = guessed_string
                    print(guessed_string + "\t\t Lives: "+ str(lives) + "\tLetters Guessed:" + str(letters))
                    
                    
                     
                else:
                    lifeLost(lives)   
                    print(guessed_string + "\t\t Lives: "+ str(lives) + "\tLetters Guessed:" + str(letters))
                    
            


    def randomWord(random_word):
        word_len = len(random_word)
        no_guesses = '_ '*word_len
        print(no_guesses + "\t Guess a letter!")
        guessedLetter(no_guesses, random_word)




    #Opening Statements

    print('Welcome to hang man!')
    print('\nThe Categories are: Animals and New Hampshire Lakes!')
    category = input("Please enter the category you would like to play!\t")


    #Category Directory
    if category.lower() == 'animals':
        random_word = random.choice(animals)
        randomWord(random_word) 

    elif category.lower() == 'lakes':

        random_word = random.choice(lakes) 
        randomWord(random_word)


    else:
        print("That's not a category you sill goose")
        hangMan()
        
hangMan()
```
